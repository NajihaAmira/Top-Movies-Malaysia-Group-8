**SCRAPING DATA MOVIES 1-50**

```python
url = 'https://www.imdb.com/search/title/?country_of_origin=MY&ref_=adv_prv'

source = requests.get(url).text
soup = bs4.BeautifulSoup(source,'html.parser')
```

```python
movie_blocks = soup.findAll('div',{'class':'lister-item-content'})
```

```python
import pandas as pd
df = pd.DataFrame(columns=['Movie Name', 'Release Year', 'Genre', 'Duration', 'IMDb Rating', 'Votes','Description'])
count = 0
for movie_block in movie_blocks:
    try:
      mname = movie_block.find('a').get_text() # Name of the movie
    except:
      mname = None
    try:
      m_reyear = int(movie_block.find('span',{'class': 'lister-item-year'}).contents[0][1:-1]) # Release year
    except:
      m_reyear = None
    try:
      m_genre = movie_block.find('span',{'class': 'genre'}).get_text() # Genre movie
    except:
      m_genre = None
    try:
      m_duration = movie_block.find('span',{'class': 'runtime'}).get_text() # Duration movie
    except:
      m_duration = None

    try:
      m_rating = float(movie_block.find('div',{'class':'inline-block ratings-imdb-rating'}).get('data-value')) #rating
    except:
      m_rating = None
    try:
      m_votes = int(movie_block.find('span',{'name':'nv'}).get('data-value')) # votes
    except:
      m_votes = None
    try:
        p_elements = movie_block.find_all('p')
        desc = p_elements[1].text # assuming the desired element is the second <p> element
    except:
        desc = None
    count += 1
    df = df.append({'Movie Name': mname, 'Release Year': m_reyear, 'Genre':m_genre, 'Duration': m_duration, 'IMDb Rating': m_rating, 'Votes':m_votes,'Description':desc}, ignore_index=True)
    

```

# **DATA CLEANING**

**Attribute : Genre**

```python
#Removing \n in the data
result_df['Genre'] = result_df['Genre'].str.replace('\n', '')
```

**Attribute: Duration**

```python
#Removing 'min' in Duration
result_df['Duration']=result_df['Duration'].str.replace('min', '')
```

```python
#Replace null values with median
median_duration = result_df['Duration'].median()
result_df['Duration'].fillna(median_duration, inplace=True)
```

```python
result_df
```

**Attribute: Votes**

```python
mean_votes = result_df['Votes'].mean()
result_df['Votes'].fillna(mean_votes, inplace=True)
```

**Attribute: Release Year**

```python
result_df['Release Year'].fillna('-', inplace=True)
```

**Attribute: IMDb Rating**

```python
#HANDLING MISSING DATA FOR RATING
print(result_df['IMDb Rating'].isnull().sum())
```

*Rating attribute contains some null values, therefore we will be replace null values with mean.*

```python
# Replace None values with mean
mean_rating = result_df['IMDb Rating'].mean()
result_df['IMDb Rating'].fillna(mean_rating, inplace=True)
```

```python
print(result_df['IMDb Rating'].isnull().sum())
```

**Attribute: Description**

```python
#HANDLING MISSING DATA FOR DESCRIPTION
print(result_df['Description'].isnull().sum())
```

*Since Description contains no null data, we can proceed.*

```python
#Removing \n in the 'Description' attribute data
result_df['Description'] = result_df['Description'].str.replace('\n', '')
#lower case for 'Description' attribute
result_df["Description"] = result_df["Description"].str.lower()

result_df
```

*Then proceed with Lemmatization for the Description attribute*

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt')
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
# Get the list of stopwords in English
stop_words = set(stopwords.words('english'))
```

```python
#Lemmatization
from nltk.stem import WordNetLemmatizer
from nltk import word_tokenize

# Create a lemmatizer
lemmatizer = WordNetLemmatizer()

# Lemmatize the "Description" column and Remove Stopwords
result_df["Description"] = result_df["Description"].apply(lambda x: " ".join([lemmatizer.lemmatize(word) for word in word_tokenize(x) if word not in stop_words]))
result_df
```

*Done cleaning the data.*

*Data visualization for description attribute*



```python
list_words = ['see', 'full', 'summary'] #These words are removed because it repeated in most descriptions
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create the word cloud object
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = list_words,
                min_font_size = 10).generate(' '.join(result_df['Description']))

# Plot the word cloud
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)

plt.show()

```

*A word cloud (also known as a tag cloud) is a data visualization technique that displays a list of words in different sizes, where the size of a word corresponds to its frequency or importance within a given text. The most frequently occurring or important words are displayed in larger font size, while less frequent or important words are displayed in smaller font size.*

*Word clouds are commonly used to visualize the content of a large body of text, such as a book, a news article, or a social media post. They can help to quickly identify the most significant topics or themes discussed in the text, and can be useful for analyzing large amounts of unstructured data.*

*In this case, we are using word cloud for visualizing the content of the description of the movies (Description attribute). As shown in the word cloud above, we can see that the largest font size word is 'story', which make sense as most movies' description would probably start with "This story is about...". We can also see 'malaysia' or 'malaysian' word in the word cloud because the website contains alot of Malaysian movies. Overall, word clouds are a useful tool for quickly exploring and understanding large amounts of unstructured text data and can be used as a starting point for more detailed analysis.*
